```python
import math
import numpy
import matplotlib.pyplot as plt
```

```python
def VanEck(power, length):

    L = 2
    VanE = [0,0]

    while L < length:
        x = len(VanE)-1
        while x > 0:
            if VanE[-1] == VanE[x-1]:     
                VanE.append((len(VanE)-x)**(len(VanE)-x))
                x = -1
            else:
                x -= 1
            if x == 0:
                VanE.append(0)
        L += 1
    
    return VanE   
```
