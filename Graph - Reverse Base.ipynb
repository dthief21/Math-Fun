```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
import datetime
```

```python
def PlotGraph(df, Title):

    fig, ax = plt.subplots()
    
    for i in range(0,len(df.index)):
        ax.plot(df.index[i], df.Terms[i], marker="o", linestyle = "", color=(0.5, 1-i/len(df.index), 1-i/len(df.index)))
    
    ax.set(xlabel='X', ylabel='Y',
           title=Title)
    ax.grid()

    plt.show()

    return
```

```python
def toBase(x, base):

    pow = 1
    baseX = ""
    
   
    while x > 0:
        rem = (x % (base))
        baseX = str(rem) + str(baseX)
        x = int((x - rem)/(base))
        pow += 1

    return baseX
```

```python
def AddTerm(df, x, base):
    
    binX = toBase(x, base)
    reverse = binX[::-1]

    term = int(binX,base) - int(reverse,base)
    
    data = [term]
    term_df = pd.DataFrame(data, columns = ['Terms'])
    df = pd.concat([df, term_df], ignore_index = True)
    
    return df
```

```python
def ReverseBase(length, base = 2):
    start = time.time()
    
    data = [1]
    df = pd.DataFrame(data, columns = ['Terms']) 
    
    while len(df.index) < length:
        df = AddTerm(df, len(df.index), base)
        
    
    PlotGraph(df,"Reverse for base {}".format(base))
    
    print(round(time.time() - start,2), "seconds")
    return
```

```python
length = int(input("How many terms?"))
for base in range(2,11):
    ReverseBase(length, base)
    
```
